```python
import numpy as np
import pandas as pd
```

```python

company_list=['AAPL','ABT','AMD','FB','IBM','INFY','KODK','MSFT','SBUX','TSLA','TTM','TWTR']

for company in company_list:
    old=f"C:\\python_files\\stock\data\\{company}.xlsx"
    new=f"C:\\python_files\\new data\\new data\\{company}.csv"
    old_df = pd.read_excel(old)
    new_df = pd.read_csv(new)
    new_df["Date"]=pd.to_datetime(new_df["Date"])
    new_df.rename(columns={'Date':'date','Open':'open','High':'high','Low':'low','Close':'close','Adj Close':'adjusted_close','Volume':'volume'},inplace=True)
    old_df = old_df.append(new_df)
    old_df.drop_duplicates(subset="date",keep="last",inplace=True)
    old_df.to_excel(old,index=False)
    print(f"{company} done")
#     print(old_df.dtypes)
#     print(new_df.dtypes)
   
    
   
# old_ds=pd.read_csv()
```
