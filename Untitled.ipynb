```python
import pandas as pd
import numpy as np
```

```python
comp = ['AAPL','ABT','AMD','FB','IBM','INFY','KODK','MSFT','SBUX','TSLA','TTM','TWTR']
# ,'ABT','AMD','FB','IBM','INFY','KODK','MSFT','SBUX','TSLA','TTM','TWTR'
for i in comp:
    old_path = f".\\data\\{i}.xlsx"
    new_path = f".\\new_data\\{i}.csv"
    old_df =pd.read_excel(old_path)
    new_df = pd.read_csv(new_path)
    new_df.rename(columns={'Date':'date','Open':'open','High':'high','Low':'low','Close':'close','Adj Close':'adjusted_close','Volume':'volume'},inplace=True)
    new_df['date'] = pd.to_datetime(new_df['date'])
    old_df = old_df.append(new_df)
    old_df.drop_duplicates(subset=['date'],inplace=True)
    old_df.to_excel(old_path,index=False)
    print(i)
```
